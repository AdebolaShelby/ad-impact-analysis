# GA 4 Import & Cleaning

```python
# Combine the current working directory with the relative path
ga4_file = os.path.join("Resources", "Traffic_acquisition.csv")

# Read the csv file with historical data from 2022 - 2024
ga4_df = pd.read_csv(ga4_file)

# Display the data table for preview
ga4_df.head(10)
```

```python
# Ensure the date column is in datetime format
ga4_df['Date'] = pd.to_datetime(ga4_df['Date'])

# Create new columns for month, year, and day
ga4_df['Year'] = ga4_df['Date'].dt.year
ga4_df['Month'] = ga4_df['Date'].dt.month
ga4_df['Day'] = ga4_df['Date'].dt.day

# Display the first 10 rows of the dataframe to verify
ga4_df.head(10)
```

# Aggregate the data by channel, year, and month

```python
# Agggregate total sessions, engaged sessions, engagement rate, average engagement time per session, event per session, 
# event count, key events, and total revenue, group by channel, year, and month


# Group by 'channel', 'Year', and 'Month' and aggregate the specified columns
aggregated_ga4_df = ga4_df.groupby(['Channel', 'Year', 'Month']).agg({
    'Sessions': 'sum',
    'Engaged sessions': 'sum',
    'Engagement rate': 'mean',
    'Average engagement time per session': 'mean',
    'Events per session': 'mean',
    'Event count': 'sum',
    'Key events': 'sum',
    'Total revenue': 'sum'
}).reset_index()

# Sort the aggregated dataframe by 'Year' and 'Month' in ascending order
aggregated_ga4_df = aggregated_ga4_df.sort_values(by=['Year', 'Month'], ascending=[True, True])

# Display the aggregated dataframe to verify results
print("Aggregated DataFrame by Year and Month:")


print(aggregated_ga4_df[aggregated_ga4_df['Month'] == 5][aggregated_ga4_df['Year'] == 2022])

# Display the first 10 rows of the aggregated dataframe
aggregated_ga4_df.head(10)
```

```python
# check for any missing values
aggregated_ga4_df.isnull().sum()
```

```python
# check for null values
aggregated_ga4_df.isnull().sum()
```

```python
# Drop the columns not needed

# drop the rows not display, referral, sms, paid social, unassigned, affiliates, and email row under the channel column
aggregated_ga4_df = aggregated_ga4_df[aggregated_ga4_df['Channel'] != 'Display']
aggregated_ga4_df = aggregated_ga4_df[aggregated_ga4_df['Channel'] != 'Email']
aggregated_ga4_df = aggregated_ga4_df[aggregated_ga4_df['Channel'] != 'Referral']
aggregated_ga4_df = aggregated_ga4_df[aggregated_ga4_df['Channel'] != 'SMS']
aggregated_ga4_df = aggregated_ga4_df[aggregated_ga4_df['Channel'] != 'Paid Social']
aggregated_ga4_df = aggregated_ga4_df[aggregated_ga4_df['Channel'] != 'Unassigned']
aggregated_ga4_df = aggregated_ga4_df[aggregated_ga4_df['Channel'] != 'Affiliates']

# Drop the Average enagagement time per session, Event per session, and Key events column
aggregated_ga4_df = aggregated_ga4_df.drop(columns=['Average engagement time per session'])
aggregated_ga4_df = aggregated_ga4_df.drop(columns=['Events per session'])
aggregated_ga4_df = aggregated_ga4_df.drop(columns=['Key events'])


# Display the first 10 rows of the dataframe
aggregated_ga4_df.head(10)
```

```python
# change the month column to the name of the month
aggregated_ga4_df['Month'] = aggregated_ga4_df['Month'].replace({
    1: 'January', 
    2: 'February', 
    3: 'March', 
    4: 'April', 
    5: 'May', 
    6: 'June', 
    7: 'July', 
    8: 'August', 
    9: 'September', 
    10: 'October', 
    11: 'November', 
    12: 'December'
})

# Display the first 10 rows of the dataframe
aggregated_ga4_df.head(10)
```

```python
# check the data types
aggregated_ga4_df.dtypes
```

```python
# save cleaned ga4 data to a csv file
aggregated_ga4_df.to_csv('Resources/cleaned_google_analytics_data.csv', index=False)
```

# Exploratory Analysis for the Year 2023 and 2024

```python
# Total sessions by channel for the years 2023 and 2024

# Create a mapping from month names to numeric values
month_mapping = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
    'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12
}

# Replace the month names with their corresponding numeric values
aggregated_ga4_df['Month'] = aggregated_ga4_df['Month'].map(month_mapping)

# Create a new column that combines the year and month into a datetime object
aggregated_ga4_df['YearMonth'] = pd.to_datetime(aggregated_ga4_df[['Year', 'Month']].assign(Day=1))

# Filter the data for the years 2023 and 2024
filtered_df = aggregated_ga4_df[(aggregated_ga4_df['Year'] >= 2023) & (aggregated_ga4_df['Year'] <= 2024)]

# Further filter the data to include only Paid Search, Direct, and Organic Search channels
filtered_df = filtered_df[filtered_df['Channel'].isin(['Paid Search', 'Direct', 'Organic Search'])]

# Generate a range of dates for the x-axis ticks
date_range = pd.date_range(start='2023-01-01', end='2024-12-31', freq='MS')

# Plot the data
plt.figure(figsize=(18, 8))
sns.lineplot(data=filtered_df, x='YearMonth', y='Sessions', hue='Channel', ci=None, marker='o')
plt.title('Sessions by Channel for the Years 2023 and 2024')
plt.xlabel('Date')
plt.ylabel('Sessions')

# Customize the x-axis tick labels to show only month and year
date_form = DateFormatter("%b %Y")
plt.gca().xaxis.set_major_formatter(date_form)

plt.xticks(date_range, rotation=45)
plt.grid()
plt.legend(title='Channel')
plt.tight_layout()
plt.show()
```

```python
# Total revenue by channels for the years 2023 and 2024

# Print unique values in the Month column before conversion
print("Unique values in the Month column before conversion:")
print(ga4_df['Month'].unique())

# Convert the Month column to string values
ga4_df['Month'] = ga4_df['Month'].astype(str)

# Combine the Year and Month columns into a single string column
ga4_df['YearMonthStr'] = ga4_df['Year'].astype(str) + '-' + ga4_df['Month']

# Print the combined YearMonthStr column
print("Combined YearMonthStr column:")
print(ga4_df['YearMonthStr'].head())

# Convert the combined YearMonthStr column to a datetime object
ga4_df['YearMonth'] = pd.to_datetime(ga4_df['YearMonthStr'], format='%Y-%m', errors='coerce')

# Print unique values in the YearMonth column after conversion
print("Unique values in the YearMonth column after conversion:")
print(ga4_df['YearMonth'].unique())

# Verify the data after creating the YearMonth column
print("Data after creating YearMonth column:")
print(ga4_df[['YearMonth', 'Channel', 'Total revenue']].head())

# Filter the data for the years 2023 and 2024
filtered_df = ga4_df[(ga4_df['YearMonth'] >= '2023-01-01') & (ga4_df['YearMonth'] <= '2024-12-31')]

# Verify the data after filtering by year
print("Data after filtering by year:")
print(filtered_df[['YearMonth', 'Channel', 'Total revenue']].head())

# Further filter the data to include only Paid Search, Direct, and Organic Search channels
filtered_df = filtered_df[filtered_df['Channel'].isin(['Paid Search', 'Direct', 'Organic Search'])]

# Verify the data after filtering by channel
print("Data after filtering by channel:")
print(filtered_df[['YearMonth', 'Channel', 'Total revenue']].head())

# Generate a range of dates for the x-axis ticks
date_range = pd.date_range(start='2023-01-01', end='2024-12-31', freq='MS')

# Plot the data for total revenue
plt.figure(figsize=(18, 8))
sns.lineplot(data=filtered_df, x='YearMonth', y='Total revenue', hue='Channel', errorbar=None, marker='o')
plt.title('Total Revenue by Channel for the Years 2023 and 2024')
plt.xlabel('Date')
plt.ylabel('Total Revenue')

# Customize the x-axis tick labels to show only month and year
date_form = DateFormatter("%b %Y")
plt.gca().xaxis.set_major_formatter(date_form)

plt.xticks(date_range, rotation=45)
plt.grid()
plt.legend(title='Channel')
plt.tight_layout()
plt.show()
```

# Correlation heatmap for the relationship between PPC Cost and all Channels

```python
# Correlation matrix heat map for the relationship bewtween cost and all the channels

# Create a DataFrame with just the data we need
analysis_df = pd.DataFrame({
    'Month_Year': merged_2023_2024_df[merged_2023_2024_df['Channel'] == 'Paid Search'].index,
    'Cost': merged_2023_2024_df[merged_2023_2024_df['Channel'] == 'Paid Search']['Cost'].values
})

# Add sessions for each channel
for channel in ['Organic Search', 'Paid Search', 'Direct', 'Cross-network', 'Organic Shopping', 
               'Organic Social', 'Organic Video', 'Paid Shopping']:
    channel_data = merged_2023_2024_df[merged_2023_2024_df['Channel'] == channel]['Sessions']
    analysis_df[f'{channel}_Sessions'] = channel_data.values

# Calculate correlation matrix
correlation_matrix = analysis_df.drop('Month_Year', axis=1).corr()

# Create heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, 
            annot=True,
            cmap='coolwarm',
            fmt='.2f',
            linewidths=0.5,
            annot_kws={'size': 10},
            center=0
)

plt.title('Channel Sessions Correlation Matrix')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Print correlations with Cost specifically
print("\nCorrelations with PPC Cost:")
cost_correlations = correlation_matrix['Cost'].sort_values(ascending=False)
for channel, correlation in cost_correlations.items():
    if channel != 'Cost':
        print(f"{channel}: {correlation:.2f}")
```

# Correlation Analysis: PPC Cost vs the three Main Channel Sessions

```python
# First, create separate series for each channel's sessions
channel_dfs = []
for channel in ['Organic Search', 'Paid Search', 'Direct']:
    channel_data = merged_2023_2024_df[merged_2023_2024_df['Channel'] == channel]['Sessions']
    channel_dfs.append(pd.Series(channel_data.values, name=channel))

# Create the DataFrame with all channels
pivot_df = pd.concat(channel_dfs, axis=1)

# Add the Cost column
pivot_df['Cost'] = merged_2023_2024_df[merged_2023_2024_df['Channel'] == 'Paid Search']['Cost'].values

# Calculate correlation matrix
correlation_matrix = pivot_df.corr()

# Create the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, 
            annot=True, 
            cmap='coolwarm', 
            fmt='.2f',
            linewidths=0.5,
            annot_kws={'size': 12},
            square=True
)

# Adjust labels
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=12)
plt.tight_layout(pad=2.0)
plt.title('Correlation Matrix: Cost vs Channel Sessions', pad=20, size=14)
plt.show()

# Print the correlation matrix
print("\nCorrelation Matrix:")
print(correlation_matrix.round(2))
```

# Correlation Coefficient Visualization

```python
# Create a DataFrame with just the data we need
analysis_df = pd.DataFrame({
    'Cost': merged_2023_2024_df[merged_2023_2024_df['Channel'] == 'Paid Search']['Cost'].values,
    'Organic_Sessions': merged_2023_2024_df[merged_2023_2024_df['Channel'] == 'Organic Search']['Sessions'].values,
    'Paid_Sessions': merged_2023_2024_df[merged_2023_2024_df['Channel'] == 'Paid Search']['Sessions'].values,
    'Direct_Sessions': merged_2023_2024_df[merged_2023_2024_df['Channel'] == 'Direct']['Sessions'].values
})

# Calculate correlations with Cost
correlations = {
    'Cost vs Organic Search': analysis_df['Cost'].corr(analysis_df['Organic_Sessions']),
    'Cost vs Paid Search': analysis_df['Cost'].corr(analysis_df['Paid_Sessions']),
    'Cost vs Direct': analysis_df['Cost'].corr(analysis_df['Direct_Sessions'])
}

# Print the correlations
print("\nCorrelations with PPC Cost:")
for relationship, correlation in correlations.items():
    print(f"{relationship}: {correlation:.2f}")

# Create a simple bar plot to visualize these correlations
plt.figure(figsize=(10, 6))
plt.bar(correlations.keys(), correlations.values())
plt.title('Correlation of PPC Cost with Channel Sessions')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Correlation Coefficient')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
